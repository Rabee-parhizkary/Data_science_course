```python
import pandas as pd
import re

#we take every word as one list and the first word will be treated as the main one
word_list = pd.Series(input().split())

#I have two cols, one is unique chars in each word, one is the count of that unique char
char_list = word_list.apply(lambda x: re.findall(r'(.)\1*', x))

#finditer gives the whole group of similar chars so we can count the length of each of them. but findall only gives the chars, not how many they were
char_count = word_list.apply(
    lambda x: [len(m.group()) for m in re.finditer(r'(.)\1*', x)]
)

is_stretchy = word_list.index.to_series().apply(
    lambda i:
        len(char_list[0]) == len(char_list[i]) and   
        char_list[0] == char_list[i] and              
        all(
            c0 == ci or (c0 >= 3 and c0 > ci)        
            for c0, ci in zip(char_count[0], char_count[i])
        )
)
data = pd.DataFrame({'words': word_list, 'unique_chars': char_list, 'char_counts': char_count, 'is_stretchy': is_stretchy})

#True = 1 and False = 0. also the first word is the main word
print(is_stretchy.sum() - 1)
data.head()
```
